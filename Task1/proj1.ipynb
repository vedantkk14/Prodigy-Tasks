```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('population_data.csv')
df.head()
```

```python
new_df = df[['state_name', 'population_male', 'population_female', 'population_total']]
```

```python
new_df.head()
```

```python
total_male = new_df['population_male'].sum()
total_female = new_df['population_female'].sum()
total_population = new_df['population_total'].sum()
print(f"Total population of Male : {total_male}")
print(f"Total population of Female : {total_female}")
print(f"Total population: {total_population}")
```

```python
type(total_male)
```

## Male vs. Female Bar chart :-

```python
from matplotlib.ticker import FuncFormatter

# First subplot
plt.bar(x=['Male', 'Female'],
            height=[total_male, total_female],
            color=['steelblue', 'lightcoral'],
            alpha=0.7,
            edgecolor='black')

formatter = FuncFormatter(lambda x, _: f'{int(x):,}')
plt.gca().yaxis.set_major_formatter(formatter)
plt.xlabel("Total number of Male and Female Population")
plt.ylabel("Population Count")
plt.title("Population Distribution")
plt.grid(axis='y', linestyle='--', alpha=0.7)


plt.tight_layout()
plt.show()
```
